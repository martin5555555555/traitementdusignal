### Estimation de la distance


Nous avons vu dans la partie preÌceÌdente, que le but premier dâ€™un radar eÌtait de deÌtecter et dâ€™estimer la distance des objets environnants. Pour cela, la meÌthode optimale est dâ€™effectuer lâ€™intercorreÌlation entre le signal recÌ§u et le signal eÌmis.

> ##### AutocorreÌlation - IntercorreÌlation
>
> On sait que lâ€™eÌnergie se conserve par transformeÌe de Fourier (theÌoreÌ€me de Parseval) :
>
> $$ \int_{-\infty}^{+\infty} |x(t)|^2 dt = \int_{-\infty}^{+\infty} |X(f)|^2 df $$
>
> On peut ainsi deÌfinir une densiteÌ spectrale dâ€™eÌnergie :  $\Phi_x (f) = |X(f)|^2$.
>
>
> Par transformeÌe de Fourier inverse, on a :
> $$ \mathcal F^{-1} [|X(f)|^2] = x(t) * x^*(-t) = \int_{-\infty}^{+\infty}  x(t)\, x^*(t-\tau) dt $$
>
>Câ€™est la fonction dâ€™autocorreÌlation, noteÌe :
> $$ R_x(\tau) = \int_{-\infty}^{+\infty}  x(t)\, x^*(t-\tau) dt $$
>
> Elle mesure la ressemblance dâ€™un signal $x(t)$ avec sa reÌplique deÌcaleÌe en temps. Cette fonction est aÌ€ symeÌtrie hermitienne et maximale en 0. 
>
> On peut geÌneÌraliser cette fonction aÌ€ lâ€™intercorreÌlation entre 2 signaux :
> $$ R_{xy}(\tau) = \int_{-\infty}^{+\infty}  x(t)\, y^*(t-\tau) dt = x(t) * y^*(-t) = \mathcal F^{-1} [X(f)Y^*(f)] $$
>
>Pour les signaux discreÌtiseÌs en temps, on utilise la formulation suivante :
> $$ R_{xy}[k] = \sum_{n = -\infty}^{+\infty} x[n] y^*[n-k] $$

> ##### Algorithme de FFT
> Lâ€™algorithme dit de Fast Fourier Transform (FFT) duÌ‚ aÌ€ Cooley et Tukey (1965) consiste aÌ€ scinder la transformeÌe de Fourier du vecteur initial de taille N en transformeÌes de Fourier de taille N/2. Dans notre application, la vitesse de lâ€™objet cible ne nous inteÌresse pas directement. Par contre, la sceÌ€ne vue par le radar peut comporter dâ€™autres eÌchos posseÌdant une vitesse nulle (on appelle ces eÌchos indeÌsirables le clutter). On peut alors exploiter lâ€™information de Doppler pour eÌliminer les cibles posseÌdant un Doppler nul. On appelle ce type de filtre, un filtre MTI (Moving Target Indicator).

> ##### Filtrage
>Un filtre $R$ est un systeÌ€me lineÌaire invariant temporellement. Il est ainsi caracteÌriseÌ par sa reÌponse impulsionnelle, $h[n] = R[\delta[n]]$. La sortie temporelle $y[n]$ sera alors le produit de convolution de cette reÌponse impulsionnelle et de lâ€™entreÌe $x[n]$:
>
> $$ y[n] = \sum_{k = -\infty}^{+\infty} x[k] h[n-k] = x * h = h * x = \sum_{k = -\infty}^{+\infty} h[k] x[n-k] $$
>
> Par transformeÌe en Z, on sait que ce produit de convolution se transforme en produit simple :
>
> $$ Y(Z) = H(Z) X(Z) $$ 
>
> On peut alors eÌgalement caracteÌriser le filtre par sa fonction de transfert : $H(Z) = TZ[h[n]]$\\
qui peut eÌ‚tre calculeÌe simplement par $H(Z) = \cfrac{Y(z)}{X(Z)}$
> Dans le cas particulier ouÌ€ $x[n] = e^{2 i \pi \tilde{f}n}$ la sortie temporelle sâ€™eÌcrit :
>
>$$ y[n] = \sum_{k = -\infty}^{+\infty} h[k] x[n-k] = e^{2 i \pi \tilde{f}n} \sum_{k = -\infty}^{+\infty} h[k] e^{-2 i \pi \tilde{f}k} = x[n] H(e^{2 i \pi \tilde f})$$
>Autrement dit, une exponentielle complexe est simplement multiplieÌe par un nombre complexe aÌ€ la traverseÌe dâ€™un systeÌ€me lineÌaire (câ€™est une fonction propre). Ce nombre complexe, nâ€™est autre que la fonction de transfert ouÌ€ lâ€™on a limiteÌ $Z$ au cercle uniteÌ. $Z(= e^{2 i \pi \tilde f})$.Ainsi, la repreÌsentation de $H(e^{2 i \pi \tilde f})$ en fonction des freÌquences permet dâ€™analyser le gain (module) et le deÌphasage (argument) subis par les composantes freÌquentielles non amorties. Câ€™est le diagramme de Bode.


Le filtre MTI le plus simple est $H_2(Z) = \cfrac{1 - z^{-1}}{2}$. On utilise classiquement ce type de filtre en
cascade dans les systeÌ€mes reÌels : $H_M(Z) = \left(\cfrac{1 - z^{-1}}{2}\right)^{M-1}$.
